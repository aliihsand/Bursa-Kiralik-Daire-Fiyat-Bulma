```python
#Gerekli Kütüphaneler
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
#Gerekli Listeler

url_list = []
price_list = []
location_list = []
propTitle = []
propValue = []

#Özelliklerin Çekilmesi
for i in range(1,25):#2 yerine sayfa sayısı gelicek.
    url = "https://www.emlakjet.com/kiralik-daire/bursa/" + str(i) + "/" #Url'nin for döngüsü ile güncellenmesi
    r = requests.get(url) # İstek atılma
    source = BeautifulSoup(r.content, 'html.parser', from_encoding='utf-8')
    
    urls = source.find_all("div",attrs={"class":"_3qUI9q"}) #Tüm verilerin bulunması
    for url in urls:
        emlak_url = "https://emlakjet.com/" + url.a.get("href") #Verilerin linklerinin bulunması
        url_list.append(emlak_url)
        print(emlak_url)
        
        r_emlak = requests.get(emlak_url)  #Laptop sayfasına bağlan
        source_emlak = BeautifulSoup(r_emlak.content,'html.parser', from_encoding='utf-8')
        
        properties = source_emlak.find_all("div",attrs={"class":"_35T4WV"}) #Tüm özellik bilgileri al
        for prop in properties:
            title = prop.find("div", class_="_1bVOdb").text.strip()
            value = prop.find_all("div", class_="_1bVOdb")[1].text.strip()
            propTitle.append(title)
            propValue.append(value)   
    
    prices = source.find_all("p",attrs={"class":"_2C5UCT"}) #Tüm fiyatları çek
    for price in prices:
        price_list.append(price.text) #Fiyatları kaydet
        
    locations = source.find_all("div",attrs={"class":"_2wVG12"}) #Tüm markları çek
    for loca in locations:
        location_list.append(loca.text) #Markaları kaydet
        
    print("Page " + str(i) + " tamamlandı.")
    
print(str(len(url_list)) + " Adet link bulundu.")
print(str(len(price_list)) + " Adet fiyat bulundu")
print(str(len(location_list)) + " Adet lokasyon verisi bulundu")
print(str(len(propTitle)) + " Adet özellik başlığı bulundu")
print(str(len(propValue)) + " Adet özellik verisi bulundu")
```

```python
#Bulunan özellik başlıklarının benzersizliklerini bulma
columns = np.array(propTitle)
columns = np.unique(columns)
```

```python
#DataFrame oluşturma
df = pd.DataFrame(columns=columns)
df["Url"] = url_list
df["Fiyat"] = price_list
df["Lokasyon"] = location_list
```

```python
#DataFram'i kullanarak bütün verileri çekmek ve sütuna yazdırmak
for i in range(0,len(url_list)): #Url sayısı kadar döndürme
    url=df["Url"].loc[i] #Satırdaki url'yi alma
    r = requests.get(url) #Linke istek atma
    source_emlak = BeautifulSoup(r.content,'html.parser',from_encoding='utf-8')
    
    properties = source_emlak.find_all("div",attrs={"class":"_35T4WV"}) #Tüm özellik bilgileri al
    for prop in properties:
            title = prop.find("div", class_="_1bVOdb").text.strip()
            value = prop.find_all("div", class_="_1bVOdb")[1].text.strip()
            
            df[title].loc[i] = value
                
    print("Veri " + str(i+1) + " içeri aktarıldı.")  
    
print("Veri aktarımı tamamlandı")
```
